##### __4.Grouping the communes by canton and saving them into separate .csv files__

[Made with the help of the example on the given page]https://stackoverflow.com/questions/46124699/splitting-a-dataframe-into-separate-csv-files

```python
for m, n in df.groupby("Canton"):
    n.to_csv("{}.csv".format(m))
```

##### __5.Computing the mean population density for each canton and ranking the cantons from most dense to least dense__

```python
df_5 = df.groupby("Canton")["Population density per km²"].mean().sort_values(ascending=False)
df_5
```

##### __6.Computing the number of communes in each canton where more than 50 percent of their populations are aged between 20 and 64 years old__

```python
df_6 = df.loc[df['20-64 years'] > 50]
df_6
```

##### __7.1.Computing the difference between the maximum and minimum elevations for each canton__

```python
df_7_1 = df.groupby("Canton")["Elevation"].apply(lambda e: e.max()- e.min())
df_7_1
```

##### __7.2.Finding the top 5 cantons that have the largest range of elevations__

```python
df_7_2 = df.groupby("Canton")["Elevation"].apply(lambda e: e.max()- e.min()).nlargest(5)
df_7_2
```

## C. Exploration: visualizations

##### __1. Horizontal bar plot that shows the top 10 populated communes__

```python
dfc1 = df1[['Commune', 'Residents']].copy()
dfc1 = dfc1.set_index('Commune')
dfc1
```

```python
dfc1.plot(kind='barh', color='blue', title="Top 10 populated communes", legend=None)
plt.xlabel("Populations")
plt.ylabel("Communes")
```

##### __2. Horizontal stacked bar chart__

```python
df1_2=df.sort_values(["Residents"], ascending=False).head(10)
dfc2 = df1_2[["Commune","Settlement area","Agricultural area", "Wooded area", "Unproductive area"]].copy()
dfc2 = dfc2.set_index('Commune')
dfc2
```

```python
dfc2.plot.barh(stacked=True, title='10 populated communes by type of area', color=("blue", "orange","green","red"))
plt.xlabel("area type in %")
plt.xlim(0,110,20)
plt.ylabel("Commune")
plt.legend(["Settlement area","Agricultural area", "Wooded area", "Unproductive area"], loc="upper right")
plt.legend(bbox_to_anchor=(1.05, 1))
#width and height setting 
plt.rcParams["figure.figsize"] = (10,8)
plt.show()
```

##### __5.The distributions of the three age groups across three language regions__

```python
#defining plotting region with 1 row and 3 columns
fig, axes = plt.subplots(1, 3)

#width and height setting 
sns.set(rc={"figure.figsize":(15, 7)}) 

#creating boxplot in each subplot
sns.boxplot(data=df, y='0-19 years', x="Language", ax=axes[0],  hue="Language").set(title='Age group 0-19 years')
sns.boxplot(data=df, y='20-64 years', x="Language", ax=axes[1],  hue="Language").set(title='Age group 20-64 years')
sns.boxplot(data=df, y='65 years or over', x="Language", ax=axes[2],  hue="Language").set(title='Age group 65 years or over')
```
